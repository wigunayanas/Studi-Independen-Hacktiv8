```python
import numpy as np
x = np.array([1,2,4,5,6])
y0 = 2 * x
y1 = 2 ** x
plt.plot(x,y0,"rx",x,y1,"bo")
plt.title("Judul Plot")
plt.xlabel('Sumbu x')
plt.ylabel("Y Sumbu")
plt.show()
```

```python
x = np.array([1,2,4,5,6])
y0 = 2 * x
y1 = 2 ** x
plt.plot(x,y0,"r-",x,y1,"b--")
plt.title("Judul Plot")
plt.xlabel('Sumbu x')
plt.ylabel("Y Sumbu")
plt.legend(['y0=2x','y1=2^x'],loc='best')
plt.show()
```

```python
kelas =['AA','BB','CC','DD','EE','FF']
jumlah_mahasiswa = [20,30,40,15,26,12]
plt.bar(kelas,jumlah_mahasiswa,color="red")
plt.title("Judul Plot")
plt.xlabel('Sumbu x')
plt.ylabel("Y Sumbu")
plt.show()
```

```python
kelas =['AA','BB','CC','DD','EE','FF']
jumlah_mahasiswa = [20,30,40,15,26,12]
plt.barh(kelas,jumlah_mahasiswa,color="red")
plt.title("Judul Plot")
plt.xlabel('Sumbu x')
plt.ylabel("Y Sumbu")
plt.show()
```

```python
posisi = ["Tim IT","Rektok","Teknisi","Dosen","Mahasiswa","Staff"]
jumlah = [20,10,10,30,70,30]
plt.pie(jumlah,autopct="%1.0f%%",labels=posisi,startangle=60)
plt.axis('equal')
plt.title("Institusi X")
plt.show()
```
